```python
import time
import matplotlib.pyplot as plt

# Interpolation Search Function
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and key >= arr[low] and key <= arr[high]:

        comparisons += 1

        # Avoid division by zero
        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((high - low) / (arr[high] - arr[low])) *
            (key - arr[low])
        )

        if arr[pos] == key:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Main Program
n = int(input("Enter number of elements: "))

# Generate uniformly distributed sorted array
arr = [i * 10 for i in range(n)]

print("\nGenerated Array:")
print(arr)

key = int(input("\nEnter element to search: "))

start_time = time.perf_counter()

position, comparisons = interpolation_search(arr, key)

end_time = time.perf_counter()

execution_time = end_time - start_time

# Display Results
if position != -1:
    print(f"\nElement found at position: {position}")
else:
    print("\nElement not found")

print("Number of comparisons:", comparisons)
print(f"Execution Time: {execution_time:.10f} seconds")

# Complexity
print("\nTime Complexity:")
print("Best Case: O(1)")
print("Average Case: O(log log n)")
print("Worst Case: O(n)")

print("\nSpace Complexity: O(1)")

# Performance Analysis for Different Input Sizes
sizes = [100, 500, 1000, 5000, 10000, 50000]
times = []

for size in sizes:
    test_arr = [i * 10 for i in range(size)]
    test_key = test_arr[size // 2]

    start = time.perf_counter()
    interpolation_search(test_arr, test_key)
    end = time.perf_counter()

    times.append(end - start)

# Plot Graph
plt.plot(sizes, times, marker='o')
plt.title("Interpolation Search Performance")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()
```
