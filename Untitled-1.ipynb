```python
import random
import time

# Interpolation Search Function
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and key >= arr[low] and key <= arr[high]:

        comparisons += 1

        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((high - low) / (arr[high] - arr[low])) *
            (key - arr[low])
        )

        if pos < low or pos > high:
            break

        if arr[pos] == key:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


print("1. Uniform Dataset")
print("2. Non-Uniform Dataset")
choice = int(input("Enter your choice: "))

print("\n1. Enter elements manually")
print("2. Generate elements randomly")
option = int(input("Choose option: "))

# Dataset Creation
if option == 1:
    n = int(input("Enter number of elements: "))
    arr = list(map(int, input("Enter elements: ").split()))
    arr.sort()

else:
    n = int(input("Enter number of elements: "))

    if choice == 1:
        # Uniform Dataset
        arr = [i * 10 for i in range(n)]

    else:
        # Non-Uniform Dataset
        arr = sorted(random.randint(1, n*n) for _ in range(n))

print("\nDataset:")
print(arr)

key = int(input("\nEnter search key: "))

start = time.perf_counter()

position, comparisons = interpolation_search(arr, key)

end = time.perf_counter()

execution_time = end - start

# Output
if position != -1:
    print("\nKey Found")
    print("Index Position:", position)
else:
    print("\nKey Not Found")

print("Number of Comparisons:", comparisons)
print(f"Execution Time: {execution_time:.10f} seconds")

# Dataset Type Information
if choice == 1:
    print("\nDataset Type: Uniform Distribution")
    print("Expected Performance: Better")
else:
    print("\nDataset Type: Non-Uniform Distribution")
    print("Expected Performance: Slower than Uniform Data")

# Complexity
print("\nTime Complexity:")
print("Best Case: O(1)")
print("Average Case: O(log log n)")
print("Worst Case: O(n)")

print("Space Complexity: O(1)")
```
